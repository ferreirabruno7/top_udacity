```python
## importação das bibliotecas de para clusterização
from sklearn.preprocessing import Imputer
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
## definição das variáveis para clusterização
X = desafio[[
    'ating_mes',
    'pontos_mes',
    'ating_acumulado',
    'pontos_acumulado',
    'acum_acumulado',
    'ating_fim_exer',
    'pontos_fim_exer',
    'acum_fim_exer'
]]


```

```python
## normalização da matriz 
X = normalize(X.as_matrix())
```

```python
## método kmeans com definição de cluster e interação
kmeans = KMeans(
    n_clusters = 9, # testar diferentes valores
    n_init = 15, 
    max_iter = 450, 
    init = 'k-means++'
)
```

```python
##treinamento do cluster
kmeans.fit(X)
```

```python
# adiciona coluna c/ cluster

desafio['cluster'] = kmeans.labels_ 
```

```python
## analise dos cluster média do atingimento das metas, perfil dos funcionários
##0 – Não atinge nenhum índices
##1 -  Atingimento geral todas as metas médias
##2 - Não atinge Metas Mensais e nem pontos acumulados apenas metas de final de exercicio
##3 - Só atinge o acumulado e metas de final de exercício
##4 - Não atinge Metas de Final de Exercício
##5 - Não atinge acumulado e pontos
##6 - Atinge apenas as metas de pontos
##7 - Atinge todas as Metas
##8 - Não atinge Metas Mensais e Acumulado

grouped = desafio.groupby('cluster')
stacked = pd.DataFrame()
for group in grouped:
    means = group[1].mean(axis = 0)
    stacked[group[0]] = means
print(stacked)
```

```python
# total de registro por cluster

desafio['cluster'].value_counts().plot.bar()
```

```python
#importação bibliotecas metodo elbow
from scipy.spatial.distance import cdist

# k means determine k
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])
 
# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distorção ')
plt.title('Método Elbow Ambev otimização K')
plt.show()
 
```

```python
## teste para prever a qual cluster um funcionário fará parte com base nos atigimentos de metas das variaveis predidoras
##funcionario 1 - cluster 3 - Só atinge o acumulado e metas de final de exercício
##funcionario 2 - cluster 2 -Não atinge Metas Mensais e nem pontos acumulados apenas metas de final de exercicio
##funcionario 3 -  cluster 3 - Só atinge o acumulado e metas de final de exercício


data = [
        [ 0, 0, 20, 20,80,60,40,80],
        [ 20, 20, 0, 0,80,80,30,90],
        [ 20, 20, 60, 60,80,60,70,100]
    ]
kmeans.predict(data)
```

```python
##avaliação do modelo de clusterização, Matriz de confusáo com base no cluster previsto para o modelo 
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import silhouette_score
print('-----------------------------------------------------------')
print('Matriz de Confusão')

print(confusion_matrix(desafio['cluster'], kmeans.labels_))
print('-----------------------------------------------------------')
print('Classificação dos cluster')

print(classification_report(desafio['cluster'], kmeans.labels_))

print('-----------------------------------------------------------')
# Avaliação de cluster com chegagem de silhuetas

print('Checagem de Silhuetas')

silhouette_avg = silhouette_score(X, kmeans.labels_, metric='euclidean', sample_size=300)
print('silhueta media:', silhouette_avg)

print('-----------------------------------------------------------')
```

```python
#variavel a ser predita - meta no final do exercicio
y = desafio['ating_fim_exer']
```

```python
#variaveis preditoras para prever final do exercicio

x = desafio[['ating_mes',
    'ating_acumulado']]
```

```python
#importacao de bibliotecas

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=101)
```

```python
#modelo de regressao
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression(normalize=True)
```

```python
#treino dos dados
lm.fit(x_train, y_train)
```

```python
#coenficientes
lm.coef_
```

```python
#predição dos dados
prediction = lm.predict(x_test)
```

```python
#gráfico de dispersão com base na predição
plt.scatter(y_test, prediction)
plt.xlabel('Y test')
plt.ylabel('Predicao Y')
```

```python
from sklearn import metrics
```

```python
#erros de previsao do modelo
print('MAE:', metrics.mean_absolute_error(y_test, prediction))
print('MSE:', metrics.mean_squared_error(y_test, prediction))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction))**0.5)
```

```python
#distribuição de gals dos erros
sns.distplot(y_test-prediction)
```
